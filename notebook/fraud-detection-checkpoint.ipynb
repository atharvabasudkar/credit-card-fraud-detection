```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Class", data=df)

plt.title("Fraud vs Non-Fraud Transactions")
plt.xticks([0,1], ["Normal", "Fraud"])

plt.show()
```

## Analyze Transaction Amounts

```python
df["Amount"].describe()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
fraud = df[df["Class"] == 1]
normal = df[df["Class"] == 0]

print("Fraud Transaction Amount Statistics")
print(fraud["Amount"].describe())

print("\nNormal Transaction Amount Statistics")
print(normal["Amount"].describe())
```

```python
plt.figure(figsize = (8,5))

sns.boxplot(x = "Class", y = "Amount", data = df)

plt.xticks([0,1], ["Normal", "Fraud"])

plt.show()
```

## Time-Based Analysis

```python
plt.figure(figsize = (10,5))

sns.histplot(df["Time"], bins = 50)

plt.title("Transaction Time Distribution")

plt.show()
```

```python
fraud_time = fraud["Time"]

plt.figure(figsize=(10,5))

sns.histplot(fraud_time, bins=50)

plt.title("Fraud Transactions Over Time")

plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.histplot(
    data=df,
    x="Time",
    hue="Class",
    bins=50,
    multiple="stack"
)

plt.title("Fraud vs Normal Transactions Over Time")
plt.show()
```

```python
df.groupby("Class")["Time"].describe()
```

## Correlation Analysis

```python
corr_matrix = df.corr()

plt.figure(figsize=(20,15))

sns.heatmap(
    corr_matrix,
    cmap="coolwarm",
    center=0
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,10))

df.corr()["Class"].sort_values().plot(kind="barh")

plt.title("Feature Correlation with Fraud Class")
plt.show()
```
